```python
# Make a column named loan_status_is_great. It should contain the integer 1 if loan_status is "Current" or "Fully Paid." 
# Else it should contain the integer 0.

def loan_status_is_great(loan_status):
  if loan_status == ('Current', 'Fully Paid'):
    return int('1')
  else:
    return int('0')
    
```

```python
loan_status_is_great('Charged Off')
```

```python
mylist = []
for loans_tatus in examples:
  mylist.append(loan_status_is_great(loans_tatus))
mylist
```
